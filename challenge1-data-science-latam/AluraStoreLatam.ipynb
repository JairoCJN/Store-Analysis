### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda_1= pd.read_csv(url)
tienda_2 = pd.read_csv(url2)
tienda_3 = pd.read_csv(url3)
tienda_4 = pd.read_csv(url4)

tienda_1.head()
```

#1. Análisis de facturación



```python
facturacion = {
    'Tienda 1': tienda_1['Precio'].sum(),
    'Tienda 2': tienda_2['Precio'].sum(),
    'Tienda 3': tienda_3['Precio'].sum(),
    'Tienda 4': tienda_4['Precio'].sum()
}
facturacion = pd.Series(facturacion)

fig, ax = plt.subplots(figsize=(8, 5))
colores = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
barras = ax.bar(facturacion.index, facturacion.values, color=colores)
ax.set_title('Facturación total por tienda')
ax.set_xlabel('Tienda')
ax.set_ylabel('Ingresos (COP)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x/1e9:.2f}B'))
ax.bar_label(barras, labels=[f'${v/1e9:.2f}B' for v in facturacion.values])
plt.tight_layout()
plt.show()
print(facturacion)
```

# 2. Ventas por categoría

```python
VPC_CONTEO = tiendas.groupby('Categoría del Producto')['Precio'].count().sort_values()

fig, ax = plt.subplots(figsize=(10, 5))
barras = ax.barh(VPC_CONTEO.index, VPC_CONTEO.values, color='steelblue')
ax.set_title('Número de ventas por categoría (todas las tiendas)')
ax.set_xlabel('Número de ventas')
ax.bar_label(barras)
plt.tight_layout()
plt.show()
```

```python
VPC_GANANCIAS = tiendas.groupby('Categoría del Producto')['Precio'].sum().sort_values()

fig, ax = plt.subplots(figsize=(10, 5))
barras = ax.barh(VPC_GANANCIAS.index, VPC_GANANCIAS.values, color='seagreen')
ax.set_title('Ingresos totales por categoría (todas las tiendas)')
ax.set_xlabel('Ingresos (COP)')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x/1e9:.1f}B'))
ax.bar_label(barras, labels=[f'${v/1e9:.2f}B' for v in VPC_GANANCIAS.values])
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
tiendas=[tienda_1,tienda_2,tienda_3,tienda_4]
n_tiendas=4
for i in range(n_tiendas):
  tiendas[i]["Tienda"]=f"Tienda {i+1}"
tiendas=pd.concat(tiendas)
tiendas
```

```python
tiendas_calificacion=tiendas.groupby("Tienda")["Calificación"].mean().round(3)
tiendas_calificacion
```

```python

valor_minimo = tiendas["Calificación"].min()
valor_maximo = tiendas["Calificación"].max()
valores = tiendas_calificacion.values
norm = mcolors.Normalize(vmin=valores.min(), vmax=valores.max())
colores = plt.cm.RdYlGn(norm(valores))
fig, ax = plt.subplots(figsize=(10,6))
barras = ax.bar(tiendas_calificacion.index, tiendas_calificacion.values, color=colores)
ax.set_title("Promedio de calificación de las tiendas")
ax.set_ylabel(f"Calificación promedio del {valor_minimo} al {valor_maximo}")
ax.set_xlabel("Tiendas")
ax.bar_label(barras)
# Anotación general debajo del gráfico
fig.text(
    0.5, -0.05,
    "No se puede afirmar que estas calificaciones sean buenas o malas. "
    "Aunque considerando que es una escala del 1 al 5 son considerablemente buenas. "
    "Lo único que podemos valorar es que la tienda 3 tiene el mejor promedio "
    "de las 4 tiendas y que la tienda que tiene peor promedio es la tienda 1",
    ha='center', fontsize=10, wrap=True
)
plt.show()
```

# 4. Productos más y menos vendidos

```python
df_productos = tiendas["Producto"].value_counts(ascending=False).rename("Frecuencia")
df_productos
```

```python
from textwrap import wrap

valor_minimo = df_productos.min()
valor_maximo = df_productos.max()
norm = mcolors.Normalize(vmin=valor_minimo, vmax=valor_maximo)
colores = plt.cm.Greens(norm(df_productos.values))
fig, ax = plt.subplots(figsize=(15,10))
barras=ax.barh(df_productos.index, df_productos.values, color=colores)
ax.set_xlabel("Numero de ventas")
ax.set_ylabel("Producto")
ax.bar_label(barras)
ax.set_title("Numero de ventas de cada producto")
fig.text(
    0.5, 0.01,
    "Que se venda menos no significa que no genere más o menos ganancias. Tal es el caso si se compara el\n"
    "N° de ventas de mesas de noche con las ventas de iPhones. Además, la diferencia entre los valores de la frecuencia no es considerable: "
    f"el valor mínimo es de {valor_minimo} y el valor máximo es de {valor_maximo}.\n"
    "Asimismo, se podrían analizar otras métricas para entender mejor el rendimiento.",
    ha='center',
    fontsize=10,
    wrap=True
)
```

# 5. Envío promedio por tienda

```python
envio_promedio = tiendas.groupby('Tienda')['Costo de envío'].mean().round(0)

fig, ax = plt.subplots(figsize=(8, 5))
colores = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
barras = ax.bar(envio_promedio.index, envio_promedio.values, color=colores)
ax.set_title('Costo de envío promedio por tienda')
ax.set_xlabel('Tienda')
ax.set_ylabel('Costo de envío promedio (COP)')
ax.bar_label(barras, fmt='${:,.0f}')
plt.tight_layout()
plt.show()
print(envio_promedio)
```

```python
resumen = pd.DataFrame({
    'Facturación': facturacion,
    'Calificación promedio': tiendas.groupby('Tienda')['Calificación'].mean().round(3),
    'Costo envío promedio': tiendas.groupby('Tienda')['Costo de envío'].mean().round(0)
})

resumen['rank_facturacion'] = resumen['Facturación'].rank(ascending=True)
resumen['rank_calificacion'] = resumen['Calificación promedio'].rank(ascending=True)
resumen['rank_envio'] = resumen['Costo envío promedio'].rank(ascending=False)
resumen['Score'] = (resumen['rank_facturacion'] * 3
                  + resumen['rank_calificacion'] * 1
                  + resumen['rank_envio'] * 1)

print('=== RESUMEN COMPARATIVO ===')
print(resumen[['Facturación','Calificación promedio','Costo envío promedio']].to_string())
print()
print('=== SCORE FINAL (menor score = candidata a vender) ===')
print(resumen[['Score']].sort_values('Score').to_string())
tienda_vender = resumen['Score'].idxmin()
print(f'\n>>> RECOMENDACIÓN: Vender la {tienda_vender} <<<')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(resumen.index, resumen['Facturación']/1e9, color=['#4C72B0','#DD8452','#55A868','#C44E52'])
axes[0].set_title('Facturación (Billones COP)')
axes[0].set_ylabel('Billones COP')

axes[1].bar(resumen.index, resumen['Calificación promedio'], color=['#4C72B0','#DD8452','#55A868','#C44E52'])
axes[1].set_title('Calificación promedio')
axes[1].set_ylim(3.9, 4.1)

axes[2].bar(resumen.index, resumen['Costo envío promedio'], color=['#4C72B0','#DD8452','#55A868','#C44E52'])
axes[2].set_title('Costo de envío promedio (COP)')

for ax in axes:
    ax.set_xlabel('Tienda')

plt.suptitle('Comparativa de métricas por tienda', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Conclusión y Recomendación

Con base en los tres indicadores analizados, la tienda candidata a ser vendida es la **Tienda 4**.La tienda 4 tiene la **menor facturación** ($1.04B COP aproximadamente), siendo el criterio de mayor peso en la decisión.


La Tienda 1, aunque tiene el mayor costo de envío y la menor calificación, compensa con la mayor facturación, por lo que conviene conservarla.
